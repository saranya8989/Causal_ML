```python
import pickle
def depickle(loc=None):
    output = []
    with open(loc,'rb') as f:
        output.append(pickle.load(f))
    return output[0]

def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
from natsort import natsorted
class graphfigure_panel2:
    def __init__(self,importantvar=None,modelvars=None):
        self.importantvar=importantvar
        self.modelvars=modelvars
        
    def _filt_vartexts(self,text=None):
        # Create a new list with all the instances the MLR model uses variables that are
        # highly partially correlated to target
        panel2prep,panel3prep = [],[]
        for ind,obj in enumerate(self.modelvars):
            if text in obj:
                if obj[0:int(len(text))]==text:
                    try: 
                        if int(obj[-1])==0:
                            panel2prep.append(obj)
                            panel3prep.append(ind)
                    except:
                        continue
                else:
                    continue
            else:
                continue
        return panel2prep,panel3prep
    
    def filt_vartexts(self):
        return [self._filt_vartexts(self.importantvar[i])[0] for i in range(len(self.importantvar))],[self._filt_vartexts(self.importantvar[i])[1] for i in range(len(self.importantvar))]
    
    def count_repeat_pandas(self,varlist_toproc=None):
        """
        Use Pandas to obtain statistics of how many times variables at a certain pressure level are used in the MLR model
        """
        count1 = pd.Series([obj.split('_')[1] for obj in varlist_toproc]).to_frame().value_counts().to_frame(name='Count')
        count1 = count1.reindex(index=natsorted(count1.index)).reset_index()
        count1.columns = ['Pressure','Count']
        return count1

def _add_nan_to_panel2(insheet=None,preslist=preslist):
    newlist = []
    for ind,obj in enumerate(preslist):
        if (obj) in [int(obj) for obj in (insheet['Pressure'])]:
            ind = [int(obj) for obj in (insheet['Pressure'])].index(obj)
            newlist.append(insheet['Count'][ind])
        else:
            newlist.append(0)
    return newlist

def _add_nan_to_panel3(timeaxis=None,proc_obj=None):
    store = np.zeros_like(timeaxis)
    proc_index = [int(obj[0]) for obj in pd.DataFrame(proc_obj).value_counts().index]
    proc_count = list(pd.DataFrame(proc_obj).value_counts().values)
    ############################################################################################
    store = []
    for ind,timeaxisobj in enumerate(timeaxis):
        if timeaxisobj in proc_index: 
            store.append(proc_count[proc_index.index(timeaxisobj)])
        else:
            store.append(0)
    return store
```

```python
from collections import Counter
class fig4:
    def __init__(self,results=None,lags=None,var_names=None):
        self.results = results
        self.lags = lags
        self.varnames = var_names
        self.multivarnames = ['vort','rhum','vvel','div','gpot','temp','outvort','outrhum','outvvel','outdiv','outgpot','outtemp']
        
    def _split_var_and_lag(self,choosevar=None,var_and_lag=None):
        return [self.varnames[obj[0]] for obj in self.lags[choosevar]],[obj[1] for obj in self.lags[choosevar]]
        
    def _get_largest_pcorra(self,choosevar=None,mlrsize=None):
        pcstable_var_laglist = list(self.results['val_min'][choosevar].keys())
        largest_pcorr = np.flipud(np.argsort(np.abs(np.asarray(list(self.results['val_min'][choosevar].values()))))[-int(mlrsize):])
        largest_pcorra = [pcstable_var_laglist[i] for i in largest_pcorr]
        
        panel1_prep_varname = [self.varnames[obj] for obj in [pcstable_var_laglist[obj][0] for obj in largest_pcorr]]
        
        simplified_panel1prep = []
        for obj in panel1_prep_varname:
            try:
                sinmultest = int(obj.split('_')[1])
                simplified_panel1prep.append(obj.split('_')[0])
            except:
                simplified_panel1prep.append(obj)
        new_vals = Counter(simplified_panel1prep).most_common()
        new_vals = new_vals[::-1]
        return panel1_prep_varname,simplified_panel1prep,new_vals
    
    def _get_mostcommon_lag(self,panel1_prep_varname=None):
        simplified_panel1prep = []
        for obj in panel1_prep_varname:
            try:
                sinmultest = int(obj.split('_')[1])
                simplified_panel1prep.append(obj.split('_')[0])
            except:
                simplified_panel1prep.append(obj)
        new_vals = Counter(simplified_panel1prep).most_common()
        new_vals = new_vals[::-1]
        return simplified_panel1prep,new_vals
    
    def _get_profile(self,new_vals=None,choosevar=None,modelvars=None):
        importantvar = []
        for a,b in new_vals:
            if a in self.multivarnames:
                importantvar.append(a)
            else:
                continue
        prec_panel2,prec_lag_panel3 = graphfigure_panel2(importantvar,modelvars).filt_vartexts()
        #del importantvar[2],prec_panel2[2]
        dict_storecount = {}
        for ind,obj in enumerate(importantvar):
            dict_storecount[obj] = graphfigure_panel2(importantvar,modelvars).count_repeat_pandas(varlist_toproc=prec_panel2[ind])
        return dict_storecount,prec_panel2,prec_lag_panel3
    
    def _produce_total(self,storestuff=None,var=None):
        fulllist = []
        for i in range(len(storestuff)):
            try:
                fulllist.append(storestuff[i]['dict_storecount'][var])
            except:
                continue
        total_counts = pd.concat(fulllist).groupby('Pressure').sum()
        df_total = pd.DataFrame({'Pressure':list(total_counts.index),'Count':list(total_counts.values)})
        return df_total
    
    def _get_timeseries_onlymulti(self,storestuffobj=None,timeaxis=np.linspace(-24,-8,17),TYPE='causal'):
        store_preppanel3 = []
        for obj in storestuffobj['prec_lag_panel3']:
            if TYPE=='causal':
                store_preppanel3.append([storestuffobj['modellags'][objs] for objs in obj])
            elif TYPE=='lag':
                store_preppanel3.append([-storestuffobj['modellags'][objs]-1 for objs in obj])
        store_panel3 = [_add_nan_to_panel3(timeaxis,obj) for obj in store_preppanel3]

        panel3_dict = {}
        for ind,obj in enumerate(store_panel3):
            panel3_dict[list(storestuffobj['dict_storecount'].keys())[ind]] = obj
        return panel3_dict
```

```python
def _produce_combine_ts(storepanel=None,varname='div'):
    tempstores = pd.concat(storepanel,axis=1)
    prep_panel3_ts = []
    for ind,obj in enumerate(list(tempstores.columns)):
        if obj==varname:
            prep_panel3_ts.append(ind)
        else:
            continue
    return tempstores.iloc[:,prep_panel3_ts].sum(axis=1)
```

```python
def make_storeP(tau_max0=None,results=None,target='v10'):
    if target=='precip':
        indexwant = 0
    elif target=='pmin':
        indexwant = 1
    elif target=='v10':
        indexwant = 2  
    storeP = []
    for i in range(tau_max0):
        storeP.append(results['p_matrix'][3:,0:3,i+1][:,int(indexwant)]) # Inputs / Targets / Lag
    return storeP

def process_forplot(inputt=None,threshold=None):#pc_alpha):
    testfilt = np.zeros_like(inputt)
    for i,j in zip(np.where(inputt>threshold)[0],np.where(inputt>threshold)[1]):
        testfilt[i,j] = 1
    return testfilt
```

```python
#function to plot multivar pcmci
def plot_2d_plinks_multivar(array=None,figsizes=None,colors=['teal','white'],\
                          cticklabels=[r'p$_{\alpha}$<0.05',r'p$_{\alpha}$>0.05'],choosetick=None,choosename=None,
                            hlineloc = None,alpha=0.3,startout=96+1,
                         title='RH: 10wspd',saveloc='/home/saranya/Downloads/plots_output/divshear_10wspdp.png',
                           alphas=[0.1,0.01]):'out2mdewtmp'
    import matplotlib
    matplotlib.rcParams['font.sans-serif'] = "DIN Alternate"
    matplotlib.rcParams['font.family'] = "sans-serif"
    
    fig,axs = plt.subplots(1,2,figsize=figsizes)
    axs[0].pcolormesh(array.transpose(),cmap=matplotlib.colors.ListedColormap(colors),alpha=alpha,edgecolors='gray',
                   linewidths=0.5)
    for obj in hlineloc:
        axs[0].axhline(y=obj,color='k',lw=2,alpha=0.9)
    #cb = plt.colorbar()
    #cb.set_ticks([0.25,0.75])
    #cb.set_ticklabels([r'p$_{\alpha}$<0.05',r'p$_{\alpha}$>0.05'])
    #for t in cb.ax.get_yticklabels():
    #    t.set_fontsize(10)
    #    t.set_weight('bold')
    
    #choosenames = var_names[3:][0:5+1]+var_names[3:][16:19+1]
    axs[0].set_yticks(choosetick, choosename, rotation=0,weight = 'normal',\
              fontsize=10.5)
    axs[0].set_xticks(np.linspace(0,16,9),[int(obj) for obj in (np.linspace(0,16,9)*3)],weight = 'normal',\
              fontsize=10.9)
    axs[0].set_title(f'pc_alpha={str(alphas[0])}',weight = 'normal',fontsize=12,loc='left')
    axs[0].set_title('TC Inner Region',weight = 'normal',fontsize=12,loc='right')
    axs[0].set_xlabel('Time Lag [Hours]',weight = 'normal',fontsize=12)
    axs[0].set_ylim(startout,0)
    axs[0].set_xlim(16,0)
    
    axs[1].pcolormesh((array.transpose())[startout:,:],
                      cmap=matplotlib.colors.ListedColormap(colors),alpha=alpha,edgecolors='gray',linewidths=0.5)
    smallhlineloc,smallyticksloc = [obj-startout for obj in hlineloc[6:]],[obj-startout for obj in choosetick[6:]]
    for obj in smallhlineloc:
        axs[1].axhline(y=obj,color='k',lw=2,alpha=0.9)
    axs[1].set_title(f'alpha_level={str(alphas[1])}',weight = 'normal',fontsize=12,loc='left')
    axs[1].set_title('TC Outer Region',weight = 'normal',fontsize=12,loc='right')
    axs[1].set_xlabel('Time Lag [Hours]',weight = 'normal',fontsize=12)
    axs[1].set_yticks(smallyticksloc)
    axs[1].set_yticklabels([])
    axs[1].set_xticks(np.linspace(0,16,9),[int(obj) for obj in (np.linspace(0,16,9)*3)],weight = 'normal',\
              fontsize=10.9)
    plt.suptitle(title,weight = 'bold',fontsize=12,horizontalalignment='center')
    plt.gca().invert_xaxis()
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(saveloc,dpi=400,facecolor='w')
    plt.show()
    return None
```
